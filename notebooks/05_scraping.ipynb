```python
def get_movie_details(tmdb_id):
    try:
        movie = movie_api.details(tmdb_id)
        credits = movie_api.credits(tmdb_id)
        
        poster = f"https://image.tmdb.org/t/p/w500{movie.poster_path}" if movie.poster_path else None
        
        director = next((c['name'] for c in credits['crew'] if c['job'] == 'Director'), None)
        actors = [c['name'] for c in credits['cast'][:5]]  # top 5 actors
        
        return poster, director, actors
    except Exception as e:
        print(f"TMDB ID {tmdb_id} error: {e}")
        return None, None, None
```

```python
import pandas as pd

# Load your cleaned CSV
top_movies = pd.read_csv("../data-movielens/top_500_movies_tmdb.csv")

# Check number of records (rows)
print(f"Number of records: {len(top_movies)}")

# Check columns
print("Columns in the file:", top_movies.columns.tolist())

# Optionally keep only the ones you want
top_movies = top_movies[['tmdbId', 'title', 'genres']]

# Verify first few rows
print(top_movies.head())
```

```python
import time

results = []

for idx, tmdb_id in enumerate(top_movies['tmdbId'], start=1):
    try:
        movie = movie_api.details(int(tmdb_id))
        if movie is None:
            print(f"{idx}/{len(top_movies)} TMDB ID {tmdb_id} not found")
            continue

        # Genres as names
        genres = [g['name'] for g in getattr(movie, "genres", [])]

        # Collect data
        results.append({
            "tmdbId": tmdb_id,
            "title": getattr(movie, "title", None),
            "genres": genres,
            "release_date": getattr(movie, "release_date", None),
            "overview": getattr(movie, "overview", None),
            "runtime": getattr(movie, "runtime", None),
            "vote_average": getattr(movie, "vote_average", None),
            "poster_path": getattr(movie, "poster_path", None)
        })

        # Optional: small delay to avoid rate limits
        time.sleep(0.2)

    except Exception as e:
        print(f"{idx}/{len(top_movies)} TMDB ID {tmdb_id} error: {e}")
```

```python
df_scraped = pd.DataFrame(results)
df_scraped.to_csv("../data-movielens/top_500_movies_scraped.csv", index=False)
print("Scraping completed! Saved to top_500_movies_scraped.csv")
```
